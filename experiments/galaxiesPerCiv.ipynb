```python
fdata = """n galaxies_per_civ
0.8333333333333334 34070322201.28847
2.0 3323267893.484986
3.6666666666666665 652061233.9318863
4.333333333333333 413375841.4406314
5.0 275674915.8134156
5.333333333333333 232287270.47914872
6.0 164633978.3128579
8.666666666666666 58102023.77315659
12.0 21615847.053520504
15.333333333333334 10380144.702367157
20.333333333333332 4480378.2323763175
"""
data = pd.read_csv(StringIO(fdata), sep=' ')
data.columns = ['Power n', '$log_{10}$ #Galaxies/GC']
```
