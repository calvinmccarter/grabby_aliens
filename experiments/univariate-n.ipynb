```python
sc = 0.5
s = 1.0
c = s / sc
randix = 0
dpi = 70
```

```python
fig = plt.figure(figsize=(7,5), dpi=dpi)
for n,color in [(1, 'r'), (5, 'purple'), (10, 'blue')]:
    fout = (
        f"{resdir}/D={D}_n={float(n)}_N={N:.2e}_s={s:.1f}_"
        f"L={L:.1f}_c={c:.5e}_r={randix}.csv"
    )
    a = pd.read_csv(fout)
    ax = (a.OriginTime / a.OriginTime.median()).plot.kde(bw_method=0.1, c=color)
ax.set_xlabel('Origin Time (rescaled $t$)')
ax.legend(['n=1', 'n=5', 'n=10'])
ax.set_xlim(0, ax.get_xlim()[1])
ax.set_title('D=3, N=1e7, s=1, L=1, c=1');
fig.savefig(f'figures-{resdir}/univariate-n-OriginTime.png');
```
