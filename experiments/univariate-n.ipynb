```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
import scipy as sp
import seaborn as sns
```

```python
D = 3 
N = 1e8
resdir = 'nsc-tuesday' #'nsc-s1' # 'nsc-s1Ne7'
s = 1.0
L = 1.0
n_perms = 5
m = 2/3
```

```python
sc = 0.5
s = 1.0
c = s / sc
randix = 0
dpi = 70
true_ns = [2, 5, 10, 20]
code_ns = [round((truen-2*m)/(1-m)) for truen in true_ns]
#code_ns = [1, 10, 16]
plot_colors = ['r', 'purple', 'blue']
plot_colors = ['#ff72a4', '#d125ff', 'darkviolet', 'mediumblue']
```

```python
fig = plt.figure(figsize=(7,5), dpi=dpi)

for truen,coden,color in list(zip(true_ns, code_ns, plot_colors)):
    fout = (
        f"{resdir}/D={D}_n={float(coden)}_N={N:.2e}_s={s:.1f}_"
        f"L={L:.1f}_c={c:.5e}_r={randix}.csv"
    )
    a = pd.read_csv(fout)
    ax = (a.OriginTime / a.OriginTime.median()).plot.kde(bw_method=0.15, c=color)
ax.set_xlabel('Origin Time (rescaled $t$)')
ax.legend([f'n = {ttn}' for ttn in true_ns])
ax.set_xlim(0, ax.get_xlim()[1])
ax.set_title(f'D={D}, N={N:.1e}, s={s}, L={L}, c={c}');
fig.savefig(f'figures-{resdir}/univariate-n-OriginTime.png');
```

```python
fig = plt.figure(figsize=(7,5), dpi=dpi)
for truen,coden,color in list(zip(true_ns, code_ns, plot_colors)):
    fout = (
        f"{resdir}/D={D}_n={float(coden)}_N={N:.2e}_s={s:.1f}_"
        f"L={L:.1f}_c={c:.5e}_r={randix}.csv"
    )
    a = pd.read_csv(fout)
    a['MinWait'] = 0.5* (a.MinArrival - a.OriginTime)
    ax = (a.MinWait / a.MinWait.median()).plot.kde(bw_method=0.1, c=color)
ax.set_xlabel('Minimum Wait Time (rescaled $t$)')
ax.legend([f'n = {ttn}' for ttn in true_ns])
ax.set_xlim(0, ax.get_xlim()[1])
ax.set_title(f'D={D}, N={N:.1e}, s={s}, L={L}, c={c}');
fig.savefig(f'figures-{resdir}/univariate-n-MinWait.png');
```
